```python
top_categories = df['category'].value_counts().head(10)

plt.figure(figsize=(12,6))
top_categories.plot(kind='bar')
plt.title("Top 10 Product Categories")
plt.xlabel("Category")
plt.ylabel("Number of Products")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

df['discount_percentage'] = df['discount_percentage'].str.replace('%','').astype(float)

df['discounted_price'] = df['discounted_price'].str.replace('₹','', regex=False).str.replace(',','', regex=False).astype(float)

df['actual_price'] = df['actual_price'].str.replace('₹','', regex=False).str.replace(',','', regex=False).astype(float)
```

```python
df['rating'].dtype
```

```python
top_products = df.sort_values(by='rating', ascending=False).head(10)

plt.figure(figsize=(12,6))
plt.bar(top_products['product_name'], top_products['rating'])
plt.title("Top 10 Highest Rated Products")
plt.xlabel("Product")
plt.ylabel("Rating")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['discount_percentage'], bins=20)
plt.title("Discount Percentage Distribution")
plt.xlabel("Discount Percentage")
plt.ylabel("Number of Products")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['actual_price'], df['discounted_price'])
plt.title("Actual Price vs Discounted Price")
plt.xlabel("Actual Price")
plt.ylabel("Discounted Price")
plt.show()
```

```python
top_reviews = df.sort_values(by='rating_count', ascending=False).head(10)

plt.figure(figsize=(12,6))
plt.bar(top_reviews['product_name'], top_reviews['rating_count'])
plt.title("Top 10 Most Reviewed Products")
plt.xlabel("Product")
plt.ylabel("Number of Ratings")
plt.xticks(rotation=90)
plt.show()
```
